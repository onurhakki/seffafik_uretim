### Download data

```python
from theohe_epias.transparency.service import WebServiceTransparency
import warnings
import pandas as pd
import holidays

pd.set_option('display.max_columns', 24)
pd.set_option('display.max_rows', 24)


warnings.simplefilter("ignore")

ws = WebServiceTransparency()
val2020 = ws.production.injection_quantity(startDate = (2020,1,1), endDate = (2020,12,31))
val2021 = ws.production.injection_quantity(startDate = (2021,1,1), endDate = (2021,12,31))
val2022 = ws.production.injection_quantity(startDate = (2022,1,1), endDate = (2022,12,31))
val2023 = ws.production.injection_quantity(startDate = (2023,1,1), endDate = (2023,12,31))
dataframe = pd.concat([val2020, val2021, val2022, val2023]).reset_index(drop=True).drop("Saat", axis = 1)


```

### Format Data

```python
def add_date_info(df):
    df = df.copy()
    df["Yıl"] = df["Tarih"].dt.year
    df["Ay"] = df["Tarih"].dt.month
    df["Gün"] = df["Tarih"].dt.day
    df["Haftanın günü"] = df["Tarih"].dt.weekday
    df["Hafta sonu"] = df["Haftanın günü"].apply(lambda x: 1 if x in [5,6] else 0)
    df["Saat"] = df["Tarih"].dt.hour
    df["Yıl günü"] = df["Tarih"].dt.day_of_year
    return df




df = add_date_info(dataframe)
### Tatiller
tr_holidays = holidays.CountryHoliday("TR")
tatiller = pd.Series(pd.date_range("2020-01-01", "2024-01-01", freq="H")).apply(tr_holidays.get).iloc[:-1].fillna(0)
df["Tatiller"] = tatiller

```
